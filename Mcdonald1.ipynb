## Category and Quantity wise Bar Chart

```python
def barplot(variable):
    var=df[variable]
    varval=var.value_counts()
    plt.figure(figsize=(10,3))
    plt.bar(varval.index,varval)
    plt.xticks(rotation=45,ha='right')
    plt.xticks(varval.index,varval.index.values)
    plt.ylabel("Friquency")
    plt.title(variable)
    plt.show()

categories=["Category"]
for a in categories:
    barplot(a)
```

# Exploratory Data Analysis (EDA)
#Analyze the distribution of calorie

```python
mean_calories = df['Calories'].mean()
median_calories = df['Calories'].median()
mode_calories = df['Calories'].mode().values[0]
std_calories = df['Calories'].std()

print(f"Mean: {mean_calories}")
print(f"Median: {median_calories}")
print(f"Mode: {mode_calories}")
print(f"Standard Deviation: {std_calories}")
```

# Distribution of Calories Counts across the Menu Items

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'], bins=20, kde=True)
plt.title('Distribution of Calorie Counts a')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

# Explore the nutritional content (e.g., fat, protein, carbohydrates) of different items.

```python
nutritional_columns = ['Total Fat', 'Protein', 'Carbohydrates']

nutritional_stats = df[nutritional_columns].describe()
print(nutritional_stats)
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[nutritional_columns])
plt.title('Distribution of Nutritional Content')
plt.ylabel('Grams')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = df[nutritional_columns].corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Nutritional Content')
plt.show()
```

# Compare nutritional characteristics of different food categories (e.g., burgers, salads, desserts).

```python
# Filter data for specific categories
selected_categories = ['Burgers', 'Salads', 'Desserts']
df_selected_categories = df[df['Category'].isin(selected_categories)]

# Display basic statistics for each category
category_stats = df_selected_categories.groupby('Category').describe()

# Create bar charts to compare nutritional characteristics
nutritional_columns = ['Total Fat', 'Protein', 'Carbohydrates']

for nutrient in nutritional_columns:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Category', y=nutrient, data=df_selected_categories)
    plt.title(f'{nutrient} Content in Burgers, Salads, and Desserts')
    plt.xlabel('Category')
    plt.ylabel(f'{nutrient} (g)')
    plt.show()
```

# Identify menu items with the highest and lowest calorie counts

```python

highest_calories = df.nlargest(5, 'Calories')  
lowest_calories = df.nsmallest(5, 'Calories')  

print("Menu Items with the Highest Calorie Counts:")
print(highest_calories[['Item', 'Calories']])

print("\nMenu Items with the Lowest Calorie Counts:")
print(lowest_calories[['Item', 'Calories']])
```
