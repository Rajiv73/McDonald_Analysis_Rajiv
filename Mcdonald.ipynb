## Category and Quantity wise Bar Chart

```python
def barplot(variable):
    var=df[variable]
    varval=var.value_counts()
    plt.figure(figsize=(10,3))
    plt.bar(varval.index,varval)
    plt.xticks(rotation=45,ha='right')
    plt.xticks(varval.index,varval.index.values)
    plt.ylabel("Friquency")
    plt.title(variable)
    plt.show()

categories=["Category"]
for a in categories:
    barplot(a)
```

## Barplot for comparing Unique_category and their Calories

```python
sns.barplot(data=df,x='Category',y='Calories')
plt.xticks(rotation=45,ha='right')
plt.title("Unique Category VS Calories")

plt.show()
```

```python
#Conclusion
#According to the count of serving Size and Calories the Chicken and Fish has the heighest Calories.
```

## Barplot for comparing Unique_category and Total Fat

```python
sns.barplot(data=df,x='Category',y='Total Fat')
plt.xticks(rotation=45,ha='right')
plt.title("Unique Category VS Total Fat")

plt.show()
```

```python
#Conclusion
#According to the count of serving Size and Fat the Chicken and Fish has the heighest fat and Beverages has lowest fat.
```

## Create a line chart for trend in protein content

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=df.index,y='Sugars',data=df,marker='o',color='blue')
plt.title("Trend in Protein Content Across Menu Items")
plt.xlabel("Menu Items(Index)")
plt.ylabel("Sugars Content (g)")
plt.show()
```

```python
#average calories count
Calories=df.groupby(["Category"])["Calories"].mean()
Sorted_Calories=Calories.sort_values(ascending=False)
print(Sorted_Calories)
```

```python
plt.figure(figsize=(7,3))
sns.swarmplot(x='Category',y='Calories',data=df,palette='colorblind')
plt.title("Total mean according to Calories")
plt.xticks(rotation=45,ha='right')
plt.xlabel("Category")
plt.ylabel("Calories")
```
